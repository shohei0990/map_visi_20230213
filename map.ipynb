```python
import requests
import json
import pandas as pd 
from urllib.parse import urlencode
```

```python
data = pd.read_csv('map_info.csv')
data.head(3)
```

```python
#パラメータリスト
api_key = 'Your API ID'
lat, lng = data['緯度'][1], data['経度'][1]
radius = 1000
keyword = "コンビニ"
language = 'ja'
minprice = 0
maxprice = 4
#fields = ['opennow']
#rankby = 'prominence'
#type = 'cafe'
```

```python
#エンドポイントURL
places_endpoint = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
#パラメータ
params = {
    "key": api_key,
    "location": f"{lat},{lng}",
    "radius": radius,
    "keyword": keyword,
    "language":language,
    #"minprice": minprice,
    #"maxprice": maxprice,
    # "fields": fields,
    # "rankby": rankby,
    # "type": type
}
```

```python
#URLエンコード
params_encoded = urlencode(params)
#リクエストURL生成
places_url = f"{places_endpoint}?{params_encoded}"

#結果取得
r = requests.get(places_url)
data = r.json()
#結果出力
print(json.dumps(data, indent=2))
```

```python
# APIレスポンスを取得
places = r.json()["results"]

df = pd.DataFrame(columns=['名称','住所','緯度','経度'])

# 各コンビニの位置情報を表示
for place in places:
    temp = pd.DataFrame( data = [[ place["name"], place["vicinity"], round(place["geometry"]["location"]["lat"], 6), round(place["geometry"]["location"]["lng"] ,6)]], columns = df.columns)
    df   = pd.concat([df, temp]) #df.append( temp,ignore_index=True ) #pd.concat([df, record])
    #df['名称'] = place["name"]
    #df['住所'] = place["vicinity"]
    #df['緯度'] = place["geometry"]["location"]["lat"]
    #df['経度'] = place["geometry"]["location"]["lng"]
    
    print("名称:", place["name"])
    print("住所:", place["vicinity"])
    print("緯度:", place["geometry"]["location"]["lat"])
    print("経度:", place["geometry"]["location"]["lng"])
    print("------------------------------")

```

```python
df = df.reset_index()
df
```

```python
from geopy.distance import geodesic
```

```python
home   = (lat, lng)
home_d = (df["緯度"][1],df["経度"][1])
```

```python
dis = geodesic(home, home_d)
```

```python
#関数化
def Map_distance(x):
    home_dis = (x["緯度"], x["経度"])
    dis = geodesic(home, home_dis).m
    return dis
```

```python
df['距離'] = df.apply(Map_distance, axis=1)
```
